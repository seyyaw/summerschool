#  Data Annotation
- Annotation is the process of `labeling data` to indicate the information it represents. This `enriches raw data` (like `texts`, `images`, or `audio`) with relevant context so that machine learning models can learn from it and make `predictions` or classifications `based on these labeled inputs`.

 ## Introduction to Data Annotation
- Importance of Data Annotation in ML
  - Data annotation crucial for optimal ML model performance
  - Takes up 80% of ML prep time: collection, cleaning, labeling
  - Proper annotation strategies enhance machine learning pipelines
  - Essential for supervised and unsupervised learning models

## Components of Data Annotation
- Key Elements for an Effective Annotation Strategy
  - Data Collection
    - Define ML project goals and data requirements
    - Ensure data variety to avoid biases and enhance model accuracy
    - Opt for appropriate data storage and formatting
  - Human Involvement
    - Human expertise vital despite automation
    - Team: engineers, scientists, annotators, managers
    - Importance of subject-matter experts for complex data

## Annotation Processes and Techniques
- Optimizing the Data Labeling Process
  - Data Labeling Phases
    - Initial training: Structured annotation according to guidelines
    - Model fine-tuning: Continuous annotation for accuracy improvement
    - Use of Human-in-the-loop (HITL) and active learning for automation synergy
  - Process Scalability
    - Iterative annotation approach with ongoing adjustments
    - Balancing manual and automated annotation efforts

## Technological Support and Strategies
- Leveraging Technology and Strategic Planning in Annotation
  - Tool Selection
    - Choose annotation tools that align with data types and ML needs
    - Automation features to handle large datasets efficiently
  - Strategic Annotation Planning
    - Establish clear objectives and guidelines for the annotation team
    - Document and secure annotation workflows to protect data integrity
    - Emphasize quality assurance and regular progress evaluations

```python
import re
def removeURL(data):
    data = ' '.join(re.sub("(@[A-Za-z0-9_]+)", "@USER", data).split())# USEE
    data =  ' '.join(re.sub('(((ftp|http|https):\/\/)|www)(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', 'URL', data).split()) # URL
    return data
```

## Summary
 - `Resources`: YouTube's entities from which you can extract information.
 -` Methods`: Functions which can be applied to the resources in order to extract their properties or modify them.
 - `List methods`: This is the method of our interest since it can be used to retrieve collections of resources and then extract information about each one of them.
   

 - `Overview section`: Where the resource is explained. The most interesting part here is the Resource representation subsection. It will be useful since from there you will select the particular properties you want to retrieve from the resources you request.
 - `list section`: Where the list method for each resource is explained.
    - In the Parameters subsection you can see the required and optional parameters that can be used in order to customize the queries.
    - In the Response subsection you can see the Response representation which includes the properties the query response has. It is important because you can also retrieve or ignore attributes of the responses.

## Example, search and save in Pandas
- About specific topics
- Get the `publish date`, `title`, `number of views`, `likes`, `dislikes`, `comments` and `favorite count`.

```python
# ሐኪም query
hakim_videos_ids = youtube.search().list(
    part="id",
    type='video',
    regionCode="ET",
    order="relevance",
    q="ሐኪም",
    maxResults=50,
    fields="items(id(videoId))"
).execute()
# ግብርና query
gibirina_videos_ids = youtube.search().list(
        part="id",
        type='video',
        regionCode="ET",
        order="relevance",
        q="ግብርና",
        maxResults=50,
        fields="items(id(videoId))"
).execute()
```

```python
hakim_info = {
    'id':[],
    'duration':[],
    'views':[],
    'likes':[],
    'favorites':[],
    'comments':[]
}
```

```python
gibirina_info = {
    'id':[],
    'duration':[],
    'views':[],
    'likes':[],
    'favorites':[],
    'comments':[]
}
```

```python
for item in hakim_videos_ids['items']:
    # Getting the id
    vidId = item['id']['videoId']
    # Getting stats of the video
    r = youtube.videos().list(
        part="statistics,contentDetails",
        id=vidId,
        fields="items(statistics," + \
                     "contentDetails(duration))"
    ).execute()
    try:
        duration = r['items'][0]['contentDetails']['duration']
        views = r['items'][0]['statistics']['viewCount']
        likes = r['items'][0]['statistics']['likeCount']
        favorites = r['items'][0]['statistics']['favoriteCount']
        comments = r['items'][0]['statistics']['commentCount']

        hakim_info['id'].append(vidId)
        hakim_info['duration'].append(duration)
        hakim_info['views'].append(views)
        hakim_info['likes'].append(likes)
        hakim_info['favorites'].append(favorites)
        hakim_info['comments'].append(comments)
    except:
        pass
```

```python
for item in gibirina_videos_ids['items']:
    vidId = item['id']['videoId']
    r = youtube.videos().list(
        part="statistics,contentDetails",
        id=vidId,
        fields="items(statistics," + \
                     "contentDetails(duration))"
    ).execute()
    try:
        duration = r['items'][0]['contentDetails']['duration']
        views = r['items'][0]['statistics']['viewCount']
        likes = r['items'][0]['statistics']['likeCount']
        favorites = r['items'][0]['statistics']['favoriteCount']
        comments = r['items'][0]['statistics']['commentCount']

        gibirina_info['id'].append(vidId)
        gibirina_info['duration'].append(duration)
        gibirina_info['views'].append(views)
        gibirina_info['likes'].append(likes)
        gibirina_info['favorites'].append(favorites)
        gibirina_info['comments'].append(comments)
    except:
        pass
```

```python
hakim_df = pd.DataFrame(data=hakim_info)
hakim_df.to_csv("hakim.csv", index=False)

gibirna_df = pd.DataFrame(data=gibirina_info)
gibirna_df.to_csv("gibirna.csv", index=False)
```

```python
def get_getcomment(vidId, count):
    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=vidId,
        maxResults=count
    )
    response = request.execute()
    for item in response['items']:
        comments.append(item['snippet']['topLevelComment']['snippet']['textDisplay'])
    return comments
```
